```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

x = np.array([5, 15, 25, 35, 45, 55]).reshape(-1, 1)
y = np.array([5,20,14,32,22,38])
print("x.shape ", x.shape)

model = LinearRegression()
model.fit(x, y)
print('intercept:', model.intercept_)
print('slope:', model.coef_)

y_hat = model.predict(x)
print('predicted response:', y_hat, sep='\n')
plt.scatter(x, y, color='black')
plt.plot(x, y_hat, color='blue', linewidth=3)
plt.xlabel('x')
plt.ylabel('y')
#plt.savefig('plots/p2predictedActual.png')
plt.show()
```

```python
x = np.array([5, 15, 25, 35, 45, 55])
y = np.array([5,20,14,32,22,38])

n = len(x)

numerator = n*np.sum(x*y) - np.sum(x)*np.sum(y)
denominator = n*np.sum(x**2) - np.sum(x)**2

b = numerator/denominator
print("slope")
print(b)

a = (np.sum(y) - b*(np.sum(x)))/n
print("intercept")
print(a)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

df = pd.read_csv('data/faithful.csv')
x = np.array(df['waiting']).reshape(-1,1)
y = np.array(df['eruptions'])

print(df.columns)

model = LinearRegression()
model.fit(x, y)

print('intercept:', model.intercept_)
print('slope:', model.coef_)

y_hat = model.predict(x)
#print('predicted response:', y_hat, sep='\n')
plt.scatter(x, y, color='black')
plt.plot(x, y_hat, color='blue', linewidth=3)
plt.xlabel('waiting')
plt.ylabel('eruptions')

print('coefficient of determination:', model.score(x, y))
```
